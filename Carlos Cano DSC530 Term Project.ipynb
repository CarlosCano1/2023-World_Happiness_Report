```python
# Covariance
```

```python
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)

    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov
```

```python
## Calculate Pearson's correlation by transforming data into arrays first
import numpy as np
x = np.asarray(Corrupt2)
y = np.asarray(Free2)
covariance = np.dot(x-x.mean(), y-y.mean()) / len(x)
correlation = covariance / math.sqrt(x.var() * y.var())
print( "Pearson's correlation: %r" %correlation)
```

```python
## Calculate Spearman's rank correlation by transforming to pandas Series
x_series = pd.Series(x)
y_series = pd.Series(y)
spearman = x_series.corr(y_series, method='spearman')
print("Spearman's correlation: %r" %spearman)
```

```python
class CorrelationPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat

    def RunModel(self):
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys
```
